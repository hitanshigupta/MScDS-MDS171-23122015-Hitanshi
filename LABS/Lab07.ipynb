My Menu Driven Restaurant

```python
import random

## Initialising the predefined menu in two-dimensional lists with item followed by price
menu = [['Veg Pizza',250],['Veg Burger',120 ],['Veg Manchurian',450],['Fried Rice',350],['Veg Noodles',350],['Lime Soda',100],['Virgin Mojito',200]]
orders = [] #an empty list to contain the ord


def display_menu(): # function definition to display the menu
    print('MENU\n')
    index = 0
    while index < len(menu): # iterate till index is less than the length
        print(index,'-',menu[index]) # printing the menu along with its index no.
        index += 1 # incrementing the index

def place_order(): #defintion to create the order

    r1 = str(random.randint(0,9)) #to store the 1st randomly generated no.
    r2 = str(random.randint(0,9)) #to store the 2nd randomly generated no.
    r3 = str(random.randint(0,9)) #to store the 3rd randomly generated no.
    r4 = str(random.randint(0,9)) #to store the 4th randomly generated no.
    r5 = str(random.randint(0,9)) #to store the 5th randomly generated no.
    orderid = r1+r2+r3+r4+r5 # concatenating the 5 randomly generated nos

    order_items = [] # defining an empty list to store item number and qty

    while True: #Iterate unless the order is completed
            
            select = input('Choose the item no. to place order for,(C to complete the order): ') #selecting the dish acc to the index no. 
        
            if select.isalpha() == True: # checking if the value in select is alphabet
                if select == 'c' or select == 'C': # if its equal to c/C, then break
                    break
                else: #otherwise
                    print('Invalid selection!!') #show msg to choose invalid selection
            else: #if not alphabet
                item_no = int(select) #storing the value of select in a variable 
                if item_no in range(len(menu)): #checking if the value is in range of the menu list
                    qty = int(input('Enter the quantity: ')) #if yes, then ask to enter qty
                    if qty > 0: #checking if the qty is greater than 0
                        order_items.append({'Itemnumber': item_no, 'Quantity': qty}) #appending the item and quantity in the list
                    else:
                        print('Quantity should be greater than 0.')
                else:
                    print('Check the item number selected!!.') #if value of select variable not in length index of menu
            
    #defining a dictionary with keys orderid and items
    order = {
        'Orderid': orderid,
        'Items': order_items,
    }
    orders.append(order) #appending the dictionary to list
    print('\nOrder',orderid,'placed successfully!')
    print('Order Details:',order,'\n')    
    print('Total Order Details:',orders,'\n')  

def display_orders(): #function to display the orders placed

    if not orders: #if no orders are placed
        print('No orders placed yet!')
        return
    
    for order in orders: #iterating the orders list
        print('Order ID:',order['Orderid']) #displaying the dictionary 
        for item in order['Items']: #iterating item wise to calculate the total
            item_no = item['Itemnumber']
            qty = item['Quantity']
            item_name = menu[item_no][0]
            item_price = menu[item_no][1]
            print(item_name, '-', qty, 'x Rs',item_price, '= Rs',qty * item_price)
        total = 0
        for item in order['Items']: #iterating item wise to calculate the total
            item_no = item['Itemnumber']
            qty = item['Quantity']
            total += menu[item_no][1] * qty
        print('Total: Rs',total)
        print()

def save_to_file(): # function defintion to save the orders in a text file 
    with open('Order_Details.txt','w') as file: #opening the file in write mode
        for order in orders:
            file.write("Order ID: " + str(order['Orderid']) + "\n")
            for item in order['Items']:
                item_no = item['Itemnumber']
                qty = item['Quantity']
                item_name = menu[item_no][0]
                item_price = menu[item_no][1]
                file.write(item_name + " - " + str(qty) + " x Rs" + str(item_price) + " = Rs" + str(qty * item_price) + "\n")
                total = 0
            for item in order['Items']:
                item_no = item['Itemnumber']
                qty = item['Quantity']
                total += menu[item_no][1] * qty
            file.write("Total: Rs" + str(total) + "\n\n")
    print("Orders saved to 'Order_Details.txt' successfully.")
    
```

```python
## MENU-DRIVEN PROGRAM

print('*'*30)
print('Welcome To The Enchanted Fork!!')

while True:
    print('*'*30)
    print('1. Place Orders')
    print('2. Display orders + Bill')
    print('3. Save Order Details to .txt file')
    print('4. Exit')
    print('*'*30)

    choice = input('Enter your choice')
    print('Your choice is:',choice,'\n')

    if int(choice) == 1:
        display_menu() #function calls
        place_order()
    elif int(choice) == 2:
        display_orders()
    elif int(choice) == 3:
        save_to_file()
    elif int(choice) == 4:
        break
    else:
        print('Invalid Selection')
```
