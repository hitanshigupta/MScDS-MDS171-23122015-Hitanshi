MATRIX MULTIPLICATION

```python
matdict = {} #defining a dictionary to store the matrices
result = [] #a list to store the result of matrix multiplication

def collectMatrix(): # general function to collect the matrices

    matrix = [] #list to store the elements in the matrix
    rows = int(input("Enter the No. of Rows: ")) #get no. of rows
    columns = int(input("Enter the No. of Columns: ")) #get no. of columns

    if rows == 0 or columns == 0: #checking if rows or columns are entered as 0
        print('Zero not possible!!')
        return

    for row in range(0,rows,1): #loop to store elements in matrix at a particular row and column
        row_val = []
        for col in range(0,columns,1):
            col_val = int(input("Enter the value for position [{},{}]".format(row,col)))
            row_val.append(col_val)
        matrix.append(row_val)

    for row in matrix: #displaying the entered matrix
        for item in row:
            print(item,end="\t")
        print()
    print()
    
    return matrix

# collecting matrices from user

def matrixcollection1(): #matrix 1

    print("Collecting Matrix 1")
    matdict["matrix 1"] = collectMatrix()
    while True:
        ch = input('Are you sure for the matrix Y/N:').strip().upper() #asking for user confirmation
        if ch == 'N':
            matdict["matrix 1"] = collectMatrix()
        elif ch == 'Y':
            break
        else:
            print('Enter valid input!!')

def matrixcollection2():
    
    print("Collecting Matrix 2") #matrix 2
    matdict["matrix 2"] = collectMatrix()
    while True:
        ch = input('Are you sure for the matrix Y/N:').strip().upper() #asking for user confirmation
        if ch == 'N':
            matdict["matrix 2"] = collectMatrix()
        elif ch == 'Y':
            break
        else:
            print('Enter valid input!!')


# function for matrix multiplication
def matrixMultiplication():
        
    if not matdict["matrix 1"] or not matdict["matrix 2"]:
        print("Error: Both matrices must be entered.")
        return

    mat_1_row = len(matdict['matrix 1'])
    mat_1_col = len(matdict['matrix 1'][0])
    mat_2_row = len(matdict['matrix 2'])
    mat_2_col = len(matdict['matrix 2'][0])
    
        
    result.clear()  
    if mat_1_col != mat_2_row: #basic condition to check matrix multiplication compatibility
        print("Error: Matrices are not compatible for multiplication.")
        return
  
    for i in range(0,mat_1_row): 
        temp_res_row = []
        for j in range(0,mat_2_col):
            tmp = 0
            for k in range(0,mat_2_col):
                tmp+=matdict['matrix 1'][i][k] * matdict['matrix 2'][k][j]
            temp_res_row.append(tmp)

        result.append(temp_res_row)

    print('The product of matrices:')
    for row in result:
        for item in row:
            print(item,end="\t")
        print()
    print()

#  function for exporting to file
def export_to_file(result):
    file = open("Matrixproduct.txt","a+") #opening file in append mode
    file.write('Matrix 1:\n') #writing first input matrix to the file
    for row in matdict['matrix 1']:
        for item in row:
            file.write(str(item) + '\t')
        file.write('\n')
    file.write('Matrix 2:\n') #writing second input matrix to the file
    for row in matdict['matrix 2']:
        for item in row:
            file.write(str(item) + '\t')
        file.write('\n')
    file.write('Result Matrix:\n') #writing the result iff multiplication was compatible
    if result:
        for row in result:
            for item in row:
                file.write(str(item) + '\t')
            file.write('\n')
    else: 
        file.write('Matrix multiplication not compatible!!')
    print('Exported to file!!')

```

```python
# menu-driven program
while True:

    print('MENU')
    print()
    print('1. Enter matrix 1')
    print('2. Enter matrix 2')
    print('3. Calculate product of the matrices')
    print('4. Export resul11t to file')
    print('5. Exit')
    print()

    choice = input('Enter your choice:')
    print('Your choice is:',choice)
    
    if int(choice) == 1:
        matrixcollection1()
    elif int(choice) == 2:
        matrixcollection2()
    elif int(choice) == 3:
        matrixMultiplication()
    elif int(choice) == 4:
        export_to_file(result)
    elif int(choice) == 5:
        break
    else:
        print('Invalid Choice!!')
        
        
```
