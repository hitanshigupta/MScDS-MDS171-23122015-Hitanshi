EXPENSE TRACKER CLASS

```python
class expense_tracker(): #defining the class 
    
    def __init__(self): # init function to define two empty lists one for expense and income 
        self.transactions = {
                'expenses': [],
                'income'  : []
        }

    def clear_transactions(self): # clear function to update the file with newly added items 
        self.transactions = {
            'expenses': [],
            'income': []
        }
        print('All transactions cleared.')

    def load_from_csv(self): # function for loading the csv file and reading it to store the transactions
        with open ('//Users//hitanshi//Desktop//GITHUB//MScDS-MDS171-23122015-Hitanshi//Lab09//lab09.csv','r') as file:
            lines = file.readlines() #reading the file line by line
        for l in lines[1:]: # iterating from the second line to avoid reading the header
            values = l.strip().split(',') #removing trailing and leading whitespaces and splitting at comma
            if len(values) ==5: #if values are total = 5
                type = values[0] # value at index 0 assigned to type variable and so on for other indexes
                category = values[1]
                cost = int(values[2])
                des = values[3]
                date = values[4]
                self.store_transaction(type,category,cost,des,date) #calling store_transaction function
            else:
                print('Check the file: some content is missing!!')
        print('Data loaded successfully from the file!!')

    def store_transaction(self,type,category,cost,des,date): #function for storing the data in dictionary
        transactions = {
            'Category' :category,
            'Cost' :cost,
            'Description' :des,
            'Date' :date
        }
        if type == 'expenses': 
            self.transactions['expenses'].append(transactions)
        else:
            self.transactions['income'].append(transactions)
        return True

    def  view_transactions(self): #function to view the transaction
        print('INCOME TRANSACTIONS')
        for i in self.transactions['income']:
            print(i)
        print('EXPENSE TRANSACTIONS')
        for i in self.transactions['expenses']:
            print(i)

    def total_incomeexpense(self): #calculating the total of incomes and expenses

        total_income = 0
        for i in self.transactions['income']:
            total_income += i['Cost']
        print('Total Income is',total_income)

        total_expenses = 0
        for i in self.transactions['expenses']:
            total_expenses += i['Cost']
        print('Total Expenses are',total_expenses)

    def export_to_file(self): # function to append new items to the file
        with open('//Users//hitanshi//Desktop//GITHUB//MScDS-MDS171-23122015-Hitanshi//Lab09//lab09.csv','a+') as file:
            for i in self.transactions['income']:
                file.write(str(i)+ '\n')
            for i in self.transactions['expenses']:
                file.write(str(i) + '\n')
            print('Exported to file successfully')

def get_input(): #function to get user inputs
    category = input('Enter Category: ')
    cost = int(input('Enter the amount: '))
    des = input('Enter description: ')
    date = input('Enter Date (DD/MM/YY): ')
    return category,cost,des,date

mytransactions = expense_tracker() #object creation
mytransactions.load_from_csv()
mytransactions.view_transactions()
mytransactions.total_incomeexpense()
```

```python
mytransactions.clear_transactions() #calling function

while True: #menu driven program
    print('...MY EXPENSE TRACKER...')
    print('1. Add Income')
    print('2. Add Expenses')
    print('3. Export to file')
    print('4. Exit')

    choice = int(input("Enter your choice: ").strip())
    print('Your choice is:',choice)

    if choice == 1:
        print("Enter the details of income")
        category,cost,des,date = get_input()
        mytransactions.store_transaction('income',category,cost,des,date)
    elif choice == 2:
        print('Enter the detai2ls of expenses')
        category,cost,des,date = get_input()
        mytransactions.store_transaction('expenses',category,cost,des,date)
    elif choice == 3:
        mytransactions.export_to_file()
    elif choice == 4:
        break
    else:
        print("In valid choice")
```
