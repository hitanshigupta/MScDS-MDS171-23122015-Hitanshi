Mix & Match Calculator

```python
import random # Importing library to use random()

items = ['Feroro Rocher','Doritos','Croissant','Thumbs Up','Ice-Cream','Lays','Kurkure','Frooty','Wafer Sticks','Peanuts'] #Items List
qty = [5,1,10,2,6,9,7,3,4,8] #Qty List
price = [500,90,200,70,100,50,30,15,60,120] #Price List

i = 1 #initial value for i i.e iteration factor

while i<=100: # to get random values for eavh list 100 times
    r1 = random.randint(0,9) #r1 to store the randomly generated value used as an index value for Items List
    r2 = random.randint(0,9) #r2 to store the randomly generated value used as an index value for Qty List
    r3 = random.randint(0,9) #r3 to store the randomly generated value used as an index value for Price List

    # Printing all the random values
    print('Random Index from Items List',r1,'\n')   
    print('Random Index from Qty List',r2,'\n')
    print('Random Index from Price List',r3,'\n')

    item = items[r1] #item to store the element at r1 index in items list
    qt = qty[r2] #qt to store the element at r2 index in qty list
    pr = price[r3] #pr to store the element at r3 index in price list

    quan = str(qt) #converting qty to string to enable writing it to the file
    cost = str(pr) #converting unit price to string to enable writing it to the file
    com = item+','+quan+','+cost #conactenating all three to get a combination
    print(com,'\n')
    
    file = open('23122015.csv','a+') # Opening the csv file in append mode
    file.write(com) #Writing the combination formed to the csv file
    file.write('\n') 
    file.close() #Closing the file

    i = i + 1 #Next iteration
```

```python
filename = '23122015.csv' #saving filename in a variable

def process(filename): #function to perform the analysis

    item_info = {} #dictionary to store information about each unique item

    with open(filename, 'r+') as file: #opening the file in read mode 
        for line in file:
            parts = line.strip().split(',')  #splitting at ',' and removing any trailing/leading spaces
            item, quantity, price = parts[0], int(parts[1]), int(parts[2])  #assigning the 3 part values to variables with type-casting to int

            if item not in item_info: #checking if an item is not in the dictionary and assigning the initial values
                item_info[item] = {
                    'Total Quantity': quantity,
                    'Minimum Price': price,
                    'Maximum Price': price,
                    'Total Cost': quantity * price
                }
            else: #if present then updating
                item_info[item]['Total Quantity'] += quantity #next qty is added 
                item_info[item]['Total Cost'] += quantity * price #total cost also added
                if price < item_info[item]['Minimum Price']: #condition to check the minimum price
                    item_info[item]['Minimum Price'] = price 
                if price > item_info[item]['Maximum Price']: #condition to check the maximum price
                    item_info[item]['Maximum Price'] = price

    return item_info #returning the dictionary
```

```python
def analysis(item_info): #function to print the analysis
    print('Item Name | Total Quantity | Minimum Price | Maximum Price | Total Cost \n') #the headings for which the data will appear
    for item, data in item_info.items():
        print(f"{item} | {data['Total Quantity']} | {data['Minimum Price']} | {data['Maximum Price']} | {data['Total Cost']}") #the analysis part
```

```python
#CALLING THE FUNCTIONS
item_info = process(filename)
analysis(item_info)
```
