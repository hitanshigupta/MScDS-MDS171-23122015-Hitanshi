```python
class ContactList:

    def __init__(self):
        self.contactlist = []

    def add_details(self,name,phone,email):

        details = {
            'Name' : name,
            'Phone' : phone,
            'Email' : email,
        }
        self.contactlist.append(details)

    def disp_details(self):
        for i in self.contactlist:
            print(i)
            print()
            print('Name:',i['Name'])
            print('Phone:',i['Phone'])
            print('Email:',i['Email'])
            print()
        print(self.contactlist)
        print()

    def edit_details(self,name):
        self.found = False
        for i in self.contactlist:
            if i['Name'] == name:
                self.found = True
                print('Editing contact',name)
                self.new_name = input('Enter the new name(Press enter if no change)').strip().title()
                self.new_phone = input('Enter the new phone no.(Press enter if no change)').strip()
                self.new_email = input('Enter the new email id.(Press enter if no change)').strip().lower()

                if self.new_name:
                    i['Name'] = self.new_name
                if self.new_phone:
                    i['Phone'] = self.new_phone
                if self.new_email:
                    i['Email'] = self.new_email
                print('Contact updated successfully!!')
        if not self.found:
            print('Contact not found!!')

    def delete_details(self,name):
        self.found = False
        for i in self.contactlist:
            if i['Name'] == name:
                self.found = True
                self.contactlist.remove(i)
                print('Contact deleted successfully!!')

        if not self.found:
            print('Contact not found!!')

    def search_details(self,name):
        self.found =  False
        self.index = 0
        for i in self.contactlist:
            if i['Name'] == name:
                self.found = True
                break
            self.index += 1

        if self.found == True:    
            print('Contact is in the list.')
            print(self.contactlist[self.index]['Name'])
            print(self.contactlist[self.index]['Phone'])
            print(self.contactlist[self.index]['Email'])
        else:
            print('Contact NOT found!')
                        
c = ContactList()

while True:

    print('MENU')
    print()
    print('1. Add details')
    print('2. Display Details')
    print('3. Edit existing details')
    print('4. Delete a contact')
    print('5. Search for a contact')
    print('6. Exit')
    print()

    choice = input('Enter your choice')
    print('Your entered choice is:',choice)

    if int(choice)== 1:
        name = input('Enter the name:').strip().title()
        phone = input('Enter the phone no.:').strip()
        email = input('Enter the email id:').strip().lower()
        print()
        c.add_details(name,phone,email)
    elif int(choice)== 2:
        c.disp_details()
    elif int(choice)== 3:
        name = input('Enter the name to be edited:').strip().title()
        c.edit_details(name)
    elif int(choice) == 4:
        name = name = input('Enter the name to be deleted:').strip().title()
        c.delete_details(name)
    elif int(choice) == 5:
        name = name = input('Enter the name to be searched:').strip().title()
        c.search_details(name)
    elif int(choice) == 6:
        break
    else:
        print('Invalid selection!!')
```
