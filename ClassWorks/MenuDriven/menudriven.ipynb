```python
namelist = []

def get_name():
    name = input('Enter the name:')
    name = name.strip().title()
    namelist.append(name)
    return name

def disp_names():
    print('*'*30)
    print('Names in the list are:')
    print('*'*30)
    for i in namelist:
        print(i)
    print('*'*30)

def search_name(name):
    name = name.strip().title()
    found = False 
    for i in namelist:
        if i == name:
            found = True
            pos = namelist.index(i) + 1
            break
    if found == True:    
        print('Name {} is in the list'.format(name), 'at position',pos)
    else:
        print('Name {} is NOT in the list'.format(name))

print('*'*30)
print('Name Management: Application')

while True:
    print('*'*30)
    print('1. Enter a Name')
    print('2. List the names')
    print('3. Search a Name')
    print('4. Exit')
    print('*'*30)

    choice = input('Enter your choice')
    print('Your choice is:',choice)

    if int(choice) == 1:
        name = get_name()
        print('Name {} added successfully!'.format(name))
    elif int(choice) == 2:
        disp_names()
    elif int(choice) == 3:
        name = input('Enter a name to be searched')
        search_name(name)
    elif int(choice) == 4:
        break
    else:
        print('Invalid Selection')
```
